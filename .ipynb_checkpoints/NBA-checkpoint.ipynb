```python
import pandas as pd
```

```python
from nba_api.stats.endpoints import playercareerstats
from nba_api.stats.static import players
nba_players = players.get_players()

lebron = [p for p in nba_players if p['full_name'] == 'LeBron James'][0]
print(lebron)
career = playercareerstats.PlayerCareerStats(player_id=lebron['id'])
career.get_data_frames()[0]
```

```python
from nba_api.stats.endpoints import playerdashboardbyyearoveryear

player_info = playerdashboardbyyearoveryear.PlayerDashboardByYearOverYear(player_id=lebron['id'], per_mode_detailed='PerGame')
player_df = player_info.get_data_frames()[1]
player_df.loc[:, ['GROUP_VALUE', 'PTS', 'REB', 'AST', 'PLUS_MINUS']]
```

### Data Collection Steps:
- For each season:
    - Get players who played in that season
    - For each player:
        - Query API for player stats and add to dataframe
    - Normalize data
    

```python
from nba_api.stats.endpoints import teamplayerdashboard
from nba_api.stats.endpoints import leaguedashteamstats

# Get list of teams that played in a season
leagueteams = leaguedashteamstats.LeagueDashTeamStats(season='2020-21')
teams = leagueteams.get_data_frames()[0]
team_list = teams.loc[:, ['TEAM_ID','TEAM_NAME']]
print(team_list)

# Get list of players that played in a season
warriors_id = teams.at[9, 'TEAM_ID']
teamplayers = teamplayerdashboard.TeamPlayerDashboard(warriors_id, season='2020-21')
players = teamplayers.get_data_frames()[1]
good_players = players[players.loc[:,'MIN'] > 800]
players_list = good_players.loc[:, ['PLAYER_ID', 'PLAYER_NAME']]
print(players_list)
```
